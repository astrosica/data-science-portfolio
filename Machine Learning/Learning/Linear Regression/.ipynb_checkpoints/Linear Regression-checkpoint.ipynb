## Predicting Credit Risk Score with Linear Regression
We will use linear regression to determine creditworthiness by predicting an individual's credit score using financial and demographic features. 

### About the Data
We will be using German Credit Data, which can be found on [UC Irvine's Machine Learning Repository](https://archive.ics.uci.edu/dataset/144/statlog+german+credit+data). The original data were designed for classification so many input features and the output variable are categorical. A numerical version of the data (`german.data-numeric`) was provided for models such as linear regression that require numerical features, however, I found that the column order was disrupted between these two files and it was very challenging to determine the proper column order, particularly for those that were one-hot encoded. Instead, I will perform the one-hot encoding myself.

This data includes a categorical target variable (`credit_score_label`) with 1 representing good credit and 2 representing bad credit. Since linear regression requires a continuous target variable, I used a large language model (LLM) to generate a synthetic credit score (`synthetic_credit_score`) that is consistent with the original categorical target variable and is correlated with the input features.

Below is a summary of the features in the data:

- **Attribute 1**: Status of existing checking account
- **Attribute 2**: Duration in months
- **Attribute 3**: Credit history
- **Attribute 4**: Purpose of the loan
- **Attribute 5**: Credit amount
- **Attribute 6**: Savings account/bonds
- **Attribute 7**: Present employment since
- **Attribute 8**: Installment rate as percentage of disposable income
- **Attribute 9**: Personal status and sex
- **Attribute 10**: Other debtors / guarantors
- **Attribute 11**: Present residence since
- **Attribute 12**: Property
- **Attribute 13**: Age in years
- **Attribute 14**: Other installment plans
- **Attribute 15**: Housing
- **Attribute 16**: Number of existing credits at this bank
- **Attribute 17**: Job
- **Attribute 18**: Number of people being liable to provide maintenance for
- **Attribute 19**: Telephone
- **Attribute 20**: Foreign worker
- **Attribute 21**: Credit risk label (original categorical variable)
- **Attribute 22**: Synthetic credit risk score (continuous target variable)

### Step 3: Encoding Features
Before we do any quantitative analysis, we'll want to make sure the features are properly encoded.

#### One-Hot Encoding
We'll start by one-hot encoding the nominal features, which are categorical features that don't have a natural order. For this we'll use Pandas' `pd.get_dummies()` function, which converts the categorical features into binary features. Using the `drop_first=True` argument drops the redundant binary column, which helps to reduce multicolinearity. I'll generate the dummy variables separately so we can easily convert them from boolean type (True or False) to integer values (1 or 0).

```python
# define nominal features for one-hot encoding
nominal_features = ["checking_status","credit_history","purpose","savings_account_bonds","personal_status_sex",
                    "other_debtors","property","other_installment_plans","housing","job"]

# create dummy variables for nominal features and convert to integer data type
df_dummies = pd.get_dummies(df[nominal_features],drop_first=True).astype(int)
encoded_nominal_features = df_dummies.columns.to_list()

# drop original nominal features from dataframe as they're now redundant
df = df.drop(columns=nominal_features)

# concatenate original dataframe with that containing the dummy variables
df = pd.concat([df, df_dummies], axis=1)
```

```python
df.head()
```

#### Ordinal Encoding
There is one ordinal feature, `employment_since`, which is a categorical feature that does have a natural order. Here, we'll use scikit-learn's `OrdinalEncoder()` function to map the ordered catorical features into ordered numerical features. We'll first have to define the order based on the file documentation. Once this feature is encoded, we'll drop the original ordinal feature (as it will be redundant) and rename the encoded feature. 

```python
# define ordinal features
ordinal_features = ["employment_since"]

# define order
employment_since_order = ["A71","A72","A73","A74","A75"]

# encode ordinal feature
ordinal_encoder = OrdinalEncoder(categories=[employment_since_order])
df[["employment_since_encoded"]] = ordinal_encoder.fit_transform(df[["employment_since"]])

# drop original ordinal feature as it's now redundant
df = df.drop(columns=["employment_since"])

# rename encoded feature
df = df.rename(columns={"employment_since_encoded":"employment_since"})
```

```python
df.head()
```

#### Binary 
Lastly, there are two binary features, `telephone` and `foreign_worker`, which have two possible string entries. Since linear regression relies on numerical features, we'll map these features into integers where 0 implies a 'False' and 1 implies a 'True'.

```python
# define binary features
binary_features = ["telephone","foreign_worker"]

# define mapping for binary features
binary_mappings = {"telephone": {"A191":0, "A192":1},
                   "foreign_worker": {"A201":1, "A202":0}}

# apply mapping for binary features
for feature in binary_features:
    df[feature] = df[feature].map(binary_mappings[feature])
```

With our features encoded, let's reorder the columns to place our target variables at the end and review our dataframe to ensure that all of the features are numerical and as expected.

```python
continuous_features = ["duration_in_month","credit_amount","installment_rate","residence_since","age","numberexisting_credits","dependents"]
df = df[continuous_features+ordinal_features+encoded_nominal_features+binary_features+target_variables]
df.head()
```

### Step 3: Feature Scaling
With our features properly encoded, it's time to scale the continuous and ordinal features such that their mean is zero and standard deviation is one.

```python
# collect features we want to scale
dont_scale_features = encoded_nominal_features+binary_features+target_variables
scale_features = df.drop(columns=dont_scale_features).columns

# scale features
scaler = StandardScaler()
df[scale_features] = scaler.fit_transform(df[scale_features])
```

```python
df.head()
```

### Step 4: Correlation Analysis
We'll now explore correlations between input features as well as correlations between input features and the target variable.

#### Step 4.1: Correlations Between Input Features
Identifying input features that are correlated with each other is essential for linear regression because this kind of multicollinearity can make the model difficult to interpret. We'll compute the correlation matrix for input features and identify cases where the Pearson correlation coefficient indicates a strong correlation ($|r|>0.8$) and drop these multicollinear features as necessary to improve model interpretability.

```python
corr_matrix =  df.drop(columns=target_variables).corr()

corr_feature_pairs = []
# iterate through indices of upper-right triangle
# of correlation matrix
for i in range(len(corr_matrix.columns)):
    for j in range(i+1,len(corr_matrix.columns)):
        r = corr_matrix.iloc[i, j]
        # correlated pairs if |r|>0.7
        if abs(r) > 0.7:
            col1 = corr_matrix.columns[i]
            col2 = corr_matrix.columns[j]
            corr_feature_pairs.append((col1,col2))

print("Pairs of correlated features:", corr_feature_pairs)
```

This reveals that `personal_status_sex_A92` is correlated with `personal_status_sex_A93` and `property_A124` is correlated with `housing_A153`. As such, I will drop the first of each pair, `personal_status_sex_A92` and `property_A124`.

```python
df = df.drop(columns=["personal_status_sex_A92","property_A124"])
```

#### Step 4.2: Correlations Between Input Features and Target Variable

Let's now explore correlations between features and the continuous target variable. Since linear regression models rely on linear relationships, a correlation analysis can help identify which features are most strongly associated with the target variable, and therefore have a greater potential impact on the model. I'll start by visualizing the correlation matrix and follow this with 2D histograms of synthetic credit scores vs input features.

```python
# correlation matrix
plt.figure(figsize=(14, 12))
sns.heatmap(df.corr(),annot=True,fmt=".2f",annot_kws={"size": 5},cmap="coolwarm",vmin=-1,vmax=1,square=True)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.title("Correlation Matrix", fontsize=16)
plt.show()
```

This reveals that a couple of features (`duration_in_month`, `credit_amount`) are negatively correlated with synthetic credit score ($r\leq-0.3$) while an additional couple of features (`checking_status_A14`, `age`) are positively correlated with synthetic credit score ($r\geq0.2$). ### Step 5: Linear Regression Modelling
Our data have been cleaned and pre-processed so it's finally time to train a linear regression model (almost)!

#### Step 5.2: Determine Which Features Have Statistically Significant Weights
We'll start by using `statsmodels` to determine which features have statistically significant weights. First we'll separate the features ($x$) from the target variable ($y$).

```python
# separate input features from target variable
X = df.drop(columns=target_variables)
y = df["synthetic_credit_score"]
```

From here, we can use `statsmodel.api.OLS` to fit an ordinary least squares (OLS) linear regression model on the design matrix ($X$), noting that the y-intercept (bias) has to be manually added to the model. This will allow us to review a summary of the results, including p-values.

```python
# manually add bias term and fit OLS linear regression model
X_with_intercept = sm.add_constant(X)
model = sm.OLS(y, X_with_intercept).fit()

# print summary
print(model.summary())
```

It is standard practice to consider a measurement as significant when $p<0.05$ (i.e., the relationship we're seeing has <5\% probability of occurring by random chance). The above results reveal many highly significant features $p=0$, including `credit_amount`, `installment_rate`, `age`, `checking_status_A14`, and `purpose_A41`. Similarly, there are several features that are revealed to be highly insignificant, including `residence_since` ($p=0.763$), `employment_since` ($p=0.871$), `purpose_A46` ($p=0.812$), `savings_account_bonds_A62` ($p=0.735$), `personal_status_sex_A94` ($p=0.989$), and `housing_A153` ($p=0.907$), suggesting that these features aren't adding meaningful value to the model.

I will therefore limit the model to features with statistically significant coefficients (weights) by considering those with $p<0.05$.

```python
# define statistically significant p-value
alpha = 0.05
pvalues = model.pvalues

# filter features using p-values
significant_features = pvalues[pvalues<alpha].index.to_list()

# filter y-intercept term
significant_features.remove('const')

# define design matrix with statistically significant features
X_significant_features = X[significant_features]

print(significant_features)
```

#### Step 5.2: Train Test Split
Next, I'll follow the standard process of separating the input features (`X`) from the target variable (`y`) and splitting the data into training (`_train`) and testing (`_test`) sets. Splitting the data into training and testing sets allows us to evaluate the model performance using an independent data set that the model hasn't seen during its training phase. This is an essential step for determining how well the model generalizes to unseen data.

```python
# split the data into training and testing sets
X_train,X_test,y_train,y_test = train_test_split(X_significant_features,y,test_size=0.2,random_state=42)
```

#### Step 5.3: Train the Linear Regression Model
Now we can finally train the model! It's hard not to appreciate how much work has gone into cleaning and pre-processing the data for the linear regression model, especially when training the model only requires two lines of code.

```python
# train the linear regression model
model = LinearRegression()
model.fit(X_train,y_train)
```

Before we investigate the model performance, let's take a look at the distribution of weights (coefficients) for the input features.

```python
# calculate coefficients (weights) and intercept (bias)
coefficients = pd.Series(model.coef_, index=X_significant_features.columns)
intercept = model.intercept_

# create bar chart of coefficients by feature
coefficients_sorted = coefficients.sort_values()

plt.figure(figsize=(10,8))
sns.barplot(y=coefficients_sorted.index,x=coefficients_sorted.values)
plt.yticks(fontsize=8)
plt.title("Statistically Significant Feature Weights")
plt.xlabel("Weight")
plt.ylabel("Feature")
plt.show()

# print results
print("Intercept (bias):", intercept)
print("\nFeature Coefficients (weights):")
print(coefficients.sort_values(ascending=False))
```

Recall that the coefficients are the weights ($w$) and the y-intercept is the bias term ($b$) in Equation 1. The magnitude of the coefficients (weights) reflects how strongly each feature impacts the model prediction while the sign reflects whether the impact is positive or negative. For example, `age` has a high positive weights ($w=18.47$), meaning that a higher age will increase the predicted synthetic credit score while a lower age will decrease it. In contrast, `credit_amount` has the most negative weight ($w=-24.58$), which means that a higher credit amount will decrease the predicted synthetic credit score while a lower credit amount will increase it.

#### Step 5.4: Test Model Performance
With the linear regression model trained, we can use it to predict the target variable using the test set and evaluate how well they match the actual target variables. We'll use `sklearn.metrics`' `mean_squared_error`, `mean_absolute_error`, and `r2_score` to quantify the model's performance.

```python
# predict the target variable using the test set
y_pred = model.predict(X_test)

# evaluate model performance
MSE  = mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)
MAE  = mean_absolute_error(y_test, y_pred)
R2   = r2_score(y_test, y_pred)

print(f"Root Mean Squared Error (RMSE): {RMSE:.2f}")
print(f"Mean Absolute Error (MAE): {MAE:.2f}")
print(f"R Score: {R2:.3f}")
```

The RMSE suggests that the mode's prediction in the synthetic credit score is, on average, off by 73.46 points while the MAE suggests that the typical prediction error is 48.37 points. These are quite large errors considering the range in credit scores. While the RMSE is more sensitive to outliers (given the nature of squared error), the MAE treats all errors equally. An $R^2$ score of 0.238 means that only 23.8% of the variance in the synthetic credit score is explainable by the model. Ideally this number is closer to 1.0, so there must be quite a lot of variability in the synthetic credit score that is unexaplainable by the model. Clearly, there is a lot of room for improvement in the linear regression model.
