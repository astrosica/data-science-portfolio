We will create a model to classify instances of an anonymized data set into one of two target classes using k-nearest neighbours (KNN).

## Import Libraries

```python
# math and visualization
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# machine learning
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix

%matplotlib inline
```

## Data

```python
df = pd.read_csv('KNN_Project_Data')
```

**Check the header, info, and summary statistics.**

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# EDA

**Use a pairplot to visualize the entire data set and use hue to segment the data into the two target classes.**

```python
sns.pairplot(df,hue='TARGET CLASS',data=df,palette='Set1')
```

# Standardize the Variables

Given the nature of this data set, we will need to standardize the features of the data to a consistent distribution. We will define the columns to be standardized, dropping the target class column because it doesn't need to be standardized.

```python
features = df.drop('TARGET CLASS', axis=1)
```

```python
scaler = StandardScaler()
scaler.fit(features)
```

**Transform the features to a scaled version.**

```python
scaled_features = scaler.transform(features)
```

**Convert the scaled features to a new dataframe.**

```python
df_scaledfeat = pd.DataFrame(data=scaled_features,columns=features.columns)
```

**Check the header of this new dataframe to make sure the scaling worked.**

```python
df_scaledfeat.head()
```

# Building the KNN Model

**Split the data into a training set and a testing set.**

```python
X = scaled_features
y = df['TARGET CLASS']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

**Train the KNN model using the training data set.**

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

# Predictions and Evaluations

**Predict values using the testing data set.**

```python
y_predict = knn.predict(X_test)
```

**Check confusion matrix and classification report for the model.**

```python
print(confusion_matrix(y_test,y_predict))
```

```python
print(classification_report(y_test,y_predict))
```

# Choosing a K Value

We will use the elbow method to choose the value of k that optimizes our model. We will create a for loop that trains KNN models with a range of k values and analyze the error rate of our models.

```python
error_rate = []

for k in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_predict = knn.predict(X_test)
    error_rate.append(np.mean(y_predict!=y_test))
```

**Plot the error rate vs. k for all KNN models.**

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,ls='--',color='blue',marker='o',markerfacecolor='red',markersize=10)
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.title('Error Rate vs. K Value')
plt.grid()
```

Looking at the plot above, we can see that the value of k=31 is a resonable value to choose that optimizes the error rate of the KNN model.

## Retrain with Optimized K Value

We will re-train a KNN model with the optimal value of k that we found above.

```python
k = 31
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train,y_train)
y_predict = knn.predict(X_test)
```

**Check confusion matrix and classification report for the optimized KNN model.**

```python
print(confusion_matrix(y_test,y_predict))
```

```python
print(classification_report(y_test,y_predict))
```

As we can see from the confusion matrix and classification report, our optimized KNN model performs better!
